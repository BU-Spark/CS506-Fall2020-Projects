```python
import pandas as pd
from collections import Counter
```

```python
rep_df = pd.read_csv('./output/all_rep_data.csv')
rep_df
```

```python
rep_df.rename(columns = {'Income Lev':'income_level', 'Tract Id':'tract_id', 'Stop Id':'stop_id','Dist Code':'dist_code','Route Ids':'route_ids','Rep Dist':'rep_dist'}, inplace = True)
rep_df
```

```python
level_0 = rep_df[rep_df['income_level'] == 0] 
level_0
```

```python
# returns a list of tract_ids, including repeats, for the given dataframe
def get_districts(dataframe): 
    all_districts = []
    for row in dataframe.itertuples():
        all_districts.append(row.rep_dist)
    return all_districts

# creates a dataframe containing each value of interest in the dataframe and the number of times it appears
def get_counts(full_list):
    occurences = Counter(full_list)
    occurences_df = pd.DataFrame.from_dict(occurences, orient='index').reset_index()
    return occurences_df
```

```python
level_0_dists = get_districts(level_0)
level_0_dists
```

```python
level_0_df = get_counts(level_0_dists)
level_0_df
```

```python
level_0_df = level_0_df.rename(columns={'index':'district', 0:'level_0'})
level_0_df
```

```python
level_1 = rep_df[rep_df['income_level'] == 1] 
level_1
```

```python
level_1_dists = get_districts(level_1)
level_1_df = get_counts(level_1_dists)
level_1_df = level_1_df.rename(columns={'index':'district', 0:'level_1'})
level_1_df
```

```python
both_levels_df = level_0_df.merge(level_1_df, how='outer', on='district')
both_levels_df
```

```python
both_levels_df = both_levels_df.fillna(0)
both_levels_df
```

```python
both_levels_df['total'] = both_levels_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
both_levels_df
```

```python
cols = ['level_0','level_1','total']
both_levels_df[cols] = both_levels_df[cols].astype(int)
both_levels_df
```

```python
both_levels_df.sort_values('total', ascending=False)
```

```python
both_levels_df['rep'] = ''
both_levels_df
```

```python
dist_df = pd.read_csv('./output/district_info.csv')
dist_df
```

```python
def get_rep(dist_info, impacted_dists):
    reps = []
    for row in impacted_dists.itertuples():
        rep = dist_info.loc[dist_info['Rep Dist'] == row.district, 'REP'].values[0]
        reps.append(rep)
    return reps
```

```python
reps = get_rep(dist_df, both_levels_df)
reps
```

```python
both_levels_df['rep'] = reps
both_levels_df
```

```python
sorted_impacted_dists = both_levels_df.sort_values('total', ascending=False)
sorted_impacted_dists
```

```python
sorted_impacted_dists.to_csv("./output/impacted_districts.csv")
```
