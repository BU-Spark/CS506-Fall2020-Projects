### 1. Draw a 0.5 mile radius circle around each stop, store as a column

```python
# This shows the units. Specifically, units are in meters.
gdf_tract_geo.crs
```

```python
def mile2meter(mile):
    conversion_factor = 0.62137119
    return mile / conversion_factor * 1000

def computeArea(radius):
    from math import pi
    return pi * radius * radius
```

```python
gdf_stop_circle = gpd.GeoDataFrame()
gdf_stop_circle[['STOP_ID', 'TOWN', 'STOP_NAME']] = gdf_stop[['STOP_ID', 'TOWN', 'STOP_NAME']]

radius = mile2meter(0.5)
gdf_stop_circle['geometry'] = gdf_stop['geometry'].buffer(radius)
gdf_stop_circle = gdf_stop_circle.set_crs('EPSG:26986')
gdf_stop_circle = gdf_stop_circle.set_geometry('geometry')
gdf_stop_circle.to_file('./output/radius.shp', driver='ESRI Shapefile')

gdf_stop_circle.head()
```

### 2. Calculate percentage of the interception between the circle and tracts
- Find overlapping tracts
- Calculate overlapping area
- Calculate proportion
    - for median income, $proportion_{ij}=\frac{overlapping\_area_{ij}}{area_{circle}}$ for tract $i$ intercepted with circle around stop $j$.
    - for population, $proportion_{ij}=\frac{overlapping\_area_{ij}}{area_{j}}$ for tract $i$ intercepted with circle around stop $j$.

```python
circle_area = computeArea(mile2meter(0.5))
circle_area
```

```python
import warnings
# ignore warnings
warnings.filterwarnings(action='ignore')

# Changed all references from TRACTCE10 to GEOID10
gdf_stop_circle['impacted_geoid'] = ""
gdf_stop_circle['proportion_2tract'] = ""
gdf_stop_circle['proportion_2circle'] = ""
for idx, row in gdf_stop_circle.iterrows():
    # tract_ids for overlapping
    overlaps = gdf_tract_geo[gdf_tract_geo['geometry'].overlaps(row['geometry'])|gdf_tract_geo['geometry'].covers(row['geometry'])|gdf_tract_geo['geometry'].within(row['geometry'])]['GEOID10'].tolist()
    # area for each overlapping
    proportion_2tract = []
    proportion_2circle = []
    for y in overlaps:
        tract_area = sum(gdf_tract_geo.loc[gdf_tract_geo['GEOID10']==y,]['SHAPE_AREA'].tolist())
        gpd_interceptions = gpd.overlay(gdf_tract_geo.loc[gdf_tract_geo.GEOID10==y,], gdf_stop_circle.loc[gdf_stop_circle.STOP_ID==row.STOP_ID,], how='intersection')
        area = sum([row.geometry.area for idx, row in gpd_interceptions.iterrows()])
        proportion_2tract.append(float(area/tract_area))
        proportion_2circle.append(float(area/circle_area))
    
    # raise error if no tracts 
    if len(overlaps)==0: 
        raise ValueError()
    
    # add columns
    gdf_stop_circle.at[idx, 'impacted_geoid'] = ','.join(overlaps)
    gdf_stop_circle.at[idx, 'proportion_2tract'] = ','.join(["{:.2f}".format(x) for x in proportion_2tract])
    gdf_stop_circle.at[idx, 'proportion_2circle'] = ','.join(["{:.2f}".format(x) for x in proportion_2circle])
gdf_stop_circle
```

```python
gdf_stop_circle.to_csv('./output/stop_radius.csv', index_label=False)
```

## 3. Calculate weighed median income for people that the stop impacts

```python
df_income_tract = gpd.read_file('./output/merged_gdf/merged_gdf.shp')
df_income_tract.head()
```

```python
df_stop_circle = pd.read_csv('./output/stop_radius.csv')
df_stop_circle.head()
```

```python
# again, changed references from TRACTCE10 to GEO_ID
df_stop_circle['income'] = -1
for idx, row in df_stop_circle.iterrows():
    tracts = str(row['impacted_geoid']).split(',')
    proportions = list(map(float, str(row['proportion_2circle']).split(',')))
    incomes = []
    for i in range(len(tracts)):
        tract = tracts[i]
        income = int(df_income_tract[df_income_tract['GEO_ID']==tract].DP03_0062E.values[0])
        if income<0:
            incomes.append(0)
            proportions[i]==0
        else:
            incomes.append(income)
    
    assert len(tracts)==len(incomes)==len(proportions)
    weighed_income = (sum([incomes[i]*proportions[i] for i in range(len(tracts))]))/sum(proportions)

    df_stop_circle.at[idx, 'income'] = weighed_income
df_stop_circle
```

Here we use the income group standard according to [Pew Research](http://www.pewsocialtrends.org/2015/12/09/the-american-middle-class-is-losing-ground/), which shows as follows:

| LEVEL | INCOME GROUP | INCOME/\$ |
| :- | :- | -: |
| 0 | Lowest income | 31,000 or less|
| 1 | Lower-middle income | 31,000 - 42,000 |
| 2 | Middle-income | 42,000 - 126,000 |
| 3 | Upper-middle income | 126,000 - 188,000 |
| 4 | Higher-income | 188,000 or more |

```python
# assign income level for each stop
income_level = []
incomes = df_stop_circle['income']
for income in incomes:
    if income<=0:
        income_level.append(-1)
    elif 0 < income < 31000:
        income_level.append(0)
    elif 31000 <= income < 42000:
        income_level.append(1)
    elif 42000 <= income < 126000:
        income_level.append(2)
    elif 126000 <= income < 188000:
        income_level.append(3)
    elif 188000 <= income:
        income_level.append(4)
df_stop_circle['income_level'] = income_level
```

```python
df_stop_circle.to_csv('./output/stop_with_weighed_level.csv', index_label=False)
```

## 4. Calculate weighed number of people impacted by each stop

**Using only the columns of interest**

- DP03_0021E = Workers 16 and over commuting to work via public transportation  
- DP03_0022E = Workers 16 and over commuting to work via walking  
- DP03_0023E = Workers 16 and over commuting to work via other means (not car)

```python
columns = ['GEO_ID','DP03_0021E','DP03_0022E','DP03_0023E']
df_tract_pop = df_income_tract[columns].astype(int)

df_tract_pop
```

```python
df_tract_pop.dtypes
```

**Calculating impacted population**  
Impacted population = the population of workers that use something other than a car to commute to work

```python
df_tract_pop['impacted_pop'] = df_tract_pop.apply(lambda x: x['DP03_0021E'] + x['DP03_0022E'] + x['DP03_0023E'], axis=1)
# renaming columns
df_tract_pop.rename(columns={'DP03_0021E': 'public_transport', 'DP03_0022E': 'walking', 'DP03_0023E':'other'}, inplace=True)
df_tract_pop
```

```python
df_stop_circle = pd.read_csv('./output/stop_with_weighed_level.csv')
df_stop_circle.head()
```

```python
# changed references from Tract ID to GEO_ID
# also adjusted dataset used to create df_tract_pop to use GEO_ID instead of Tract ID
df_stop_circle['impacted_pop'] = 0
for idx, row in df_stop_circle.iterrows():
    tracts = str(row['impacted_geoid']).split(',')
    proportions = list(map(float, str(row['proportion_2tract']).split(',')))
    impacted_pops = []
    for tract in tracts:
        impacted_pop = int(df_tract_pop[df_tract_pop.GEO_ID==int(tract)].impacted_pop.values[0])
        impacted_pops.append(impacted_pop)
    weighed_pop = sum([impacted_pops[i]*proportions[i] for i in range(len(tracts))])

    df_stop_circle.at[idx, 'impacted_pop'] = weighed_pop
df_stop_circle
```

# Most Impacted Routes and Districts Calculations

```python
import pandas as pd
import geopandas as gpd
from collections import Counter
```

### Route Calculations

```python
stops_df = pd.read_csv('./output/stops.csv')
stops_df
```

```python
def get_income_level(gdf, level):
    # takes as inputs a geodataframe and an int to specify the income level, returns a geodataframe with
    # stops of that given income level
    return gdf[gdf['income_lev'] == level]

# returns a list of districts, including repeats, for the given dataframe
def get_districts_list(df): 
    all_districts = []
    for row in df.itertuples():
        all_districts.append(row.rep_dist)
    return all_districts

# returns a list of route_ids, including repeats, for the given dataframe
def get_route_ids_list(df): 
    all_route_ids = []
    for row in df.itertuples():
        row_route_ids = row.route_ids
        if pd.notna(row_route_ids):
            route_id_list = row_route_ids.split(',')
            for route_id in route_id_list:
                all_route_ids.append(route_id)
    return all_route_ids

# creates a dataframe containing each value in the dataframe and the number of times it appears
def get_counts(full_list):
    occurences = Counter(full_list)
    occurences_df = pd.DataFrame.from_dict(occurences, orient='index').reset_index()
    return occurences_df
```

```python
level_0 = get_income_level(stops_df,0)
level_0
```

```python
level_0_ids_list = get_route_ids_list(level_0)
```

```python
level_0_df = get_counts(level_0_ids_list)
level_0_df = level_0_df.rename(columns={'index':'route_id', 0:'level_0'})
level_0_df
```

```python
# Repeating for income level 1
level_1 = get_income_level(stops_df,1)
level_1_ids_list = get_route_ids_list(level_1)
level_1_df = get_counts(level_1_ids_list)
level_1_df = level_1_df.rename(columns={'index':'route_id', 0:'level_1'})
```

```python
# combining findings for level 1 and level 0
both_levels_df = level_0_df.merge(level_1_df, how='outer', on='route_id')
both_levels_df = both_levels_df.fillna(0)
both_levels_df['total'] = both_levels_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
cols = ['level_0','level_1','total']
both_levels_df[cols] = both_levels_df[cols].astype(int)
both_levels_df
```

Getting route-level population numbers by summing the populations of stops along a route

```python
# calculates the impacted_pop for each route depending on the stops it passes through
def get_pop():
    for row in stops_df.itertuples():
        row_route_ids = row.route_ids
        if pd.notna(row_route_ids):
            route_id_list = row_route_ids.split(',')
            for route in route_id_list:
                both_levels_df.loc[both_levels_df.route_id == route, "impacted_p"] += row.impacted_p
    return
```

```python
both_levels_df['impacted_p'] = 0
get_pop()
both_levels_df
```

```python
both_levels_df.to_csv("./output/weighted_route_info_FINAL.csv", index=False)
```

```python
sorted_df = both_levels_df.sort_values(by='total', ascending=False)
sorted_df
```

```python
rep_df = pd.read_csv('./output/all_rep_data_FINAL.csv')
rep_df.rename(columns = {'Income Lev':'income_lev', 'Stop Id':'stop_id','Dist Code':'dist_code','Route Ids':'route_ids','Rep Dist':'rep_dist'}, inplace = True)
rep_df
# Null values are districts that do not contain any bus stops
```

```python
level_0 = get_income_level(rep_df, 0)
level_0_dists = get_districts_list(level_0)
level_0_df = get_counts(level_0_dists)
level_0_df = level_0_df.rename(columns={'index':'district', 0:'level_0'})

level_1 = get_income_level(rep_df, 1)
level_1_dists = get_districts_list(level_1)
level_1_df = get_counts(level_1_dists)
level_1_df = level_1_df.rename(columns={'index':'district', 0:'level_1'})
```

```python
both_levels_df = level_0_df.merge(level_1_df, how='outer', on='district')
both_levels_df = both_levels_df.fillna(0)
both_levels_df['total'] = both_levels_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
cols = ['level_0','level_1','total']
both_levels_df[cols] = both_levels_df[cols].astype(int)
```

```python
dist_df = pd.read_csv('./output/district_info.csv')
```

```python
def get_rep(dist_info, impacted_dists):
    reps = []
    for row in impacted_dists.itertuples():
        rep = dist_info.loc[dist_info['Rep Dist'] == row.district, 'REP'].values[0]
        reps.append(rep)
    return reps
```

### Getting impacted towns

```python
town_count = stops_df.TOWN.value_counts()
town_count_df = town_count.rename_axis('town').reset_index(name='counts')
town_count_df
```

```python
# returns a list of towns, including repeats, for the given dataframe
def get_town(dataframe): 
    all_towns = []
    for row in dataframe.itertuples():
        all_towns.append(row.Town)
    return all_towns
```

```python
level_0_towns = get_town(level_0)
level_0_town_df = get_counts(level_0_towns)
level_0_town_df = level_0_town_df.rename(columns={'index':'town', 0:'level_0'})

level_1_towns = get_town(level_1)
level_1_town_df = get_counts(level_1_towns)
level_1_town_df = level_1_town_df.rename(columns={'index':'town', 0:'level_1'})
```

```python
all_towns_df = level_0_town_df.merge(level_1_town_df, how='outer', on='town')
all_towns_df = all_towns_df.fillna(0)
all_towns_df['total'] = all_towns_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
cols = ['level_0','level_1','total']
all_towns_df[cols] = all_towns_df[cols].astype(int)
all_towns_df
```

```python
def town_percentage(town_totals, town_low_income):
    percent_dict = {}
    for town in town_low_income.town:
        low_income_stops = town_low_income.loc[town_low_income['town'] == town, 'total'].values[0] # number of low-income stops
        all_stops = town_totals.loc[town_totals['town'] == town, 'counts'].values[0] # number of stops in total
        percent_low_income = low_income_stops/all_stops*100 # percentage of stops that are low-income for that town
        percent_dict[town] = percent_low_income
    return percent_dict
```

```python
town_percents = town_percentage(town_count_df, all_towns_df)
town_percents
```
