```python
# imports
import pandas as pd
import numpy as np
```

```python
# US Census data
table = pd.read_csv('./data/census_tract/census/ACSDP5Y2018.DP03_data_with_overlays_2020-11-12T115259.csv', low_memory=False)
```

```python
table
```

# Using only the columns of interest
DP03_0021E = Workers 16 and over commuting to work via public transportation
DP03_0022E = Workers 16 and over commuting to work via walking
DP03_0023E = Workers 16 and over commuting to work via other means (not car)
DP03_0026E = Civilian employed population 16 years and over

```python
columns = ['GEO_ID','NAME','DP03_0021E','DP03_0022E','DP03_0023E','DP03_0026E','DP03_0062E']
df = table[columns]
```

```python
df
```

```python
# creating a new column for tract id, which is the last 6 numbers of GEO_ID
df['GEO_ID'] = df['GEO_ID'].apply(lambda x: x[9:])
```

```python
df
```

```python
df.dtypes
```

```python
# removing the first row containing the descriptions for each column
df = df.iloc[1:]
df
```

```python
# Converting population columns to integers for calculations
pop_columns = ['DP03_0021E','DP03_0022E','DP03_0023E','DP03_0026E']
df[pop_columns] = df[pop_columns].astype(int)
df.dtypes
```

# Calculating impacted population 
Impacted population = the population of workers that use something other than a car to commute to work

```python
df['impacted_pop'] = df.apply(lambda x: x['DP03_0021E'] + x['DP03_0022E'] + x['DP03_0023E'], axis=1)
df
```

```python
# this column had missing values, represented by a -
df['DP03_0062E'] = df['DP03_0062E'].replace('-', '0')
```

```python
# IGNORE - the issue was solved in the previous cell
df['DP03_0062E'] = df['DP03_0062E'].astype(int)
df.dtypes
```

```python
# investigating the previous cell's issue
contain_values = df[df['DP03_0062E'].str.contains('-')]
print(contain_values)
```

```python
# done after removing the null values represented by a dash.
contain_values = df[df['DP03_0062E'].str.contains('-')]
print(contain_values)
```

```python
df.iloc[0,:]
```

```python
# a dataset we worked with previously showed that the tract with tract_id = 820400 had an income of 2499
# that dataset can be found here: https://api.census.gov/data/2018/acs/acs5?get=B19013_001E&for=tract:*&in=state:25
# this was used in Deliverable 1's calculations
df['DP03_0062E'] = df['DP03_0062E'].replace('2,500-', '2499')
```

```python
contain_values = df[df['DP03_0062E'].str.contains('-')]
print(contain_values)
```

```python
# renaming columns
df = df.rename(columns={'DP03_0021E': 'public_transport', 'DP03_0022E': 'walking', 'DP03_0023E':'other', 'DP03_0026E':'total_employed', 'DP03_0062E':'income'})
```

```python
df
```

```python
cols = ['GEO_ID','public_transport','walking','other','total_employed','impacted_pop']
final_df = df[cols]
final_df
```

```python
final_df.to_csv("./output/tracts_with_population.csv")
```
