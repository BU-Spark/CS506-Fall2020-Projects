The following functions are based off the functions found here: https://github.com/kennyalexlin/CS506-Fall2020-Projects/blob/junli/transit_equity_team1/programs/Transit_Equity.ipynb

get_pop_df_dict() creates a dictionary with tracts from pop_df's tract_id as keys and entries from the specified columns as values corresponding to the correct tract.

For a given attribute, get_value_for_tract() assigns each stop in stops_df the attribute of the tract it's located in.

```python
def get_pop_df_dict(column_name):
    pop_df_dict = {}
    
    needed_nums = pop_df[column_name]
    tracts = [str(x).zfill(6) for x in pop_df['tract_id']]
    for i in range(len(pop_df)):
        pop_df_dict[tracts[i]] = needed_nums[i]
    return pop_df_dict

def get_value_for_tract(dictionary):
    values = []
    tracts = [str(x).zfill(6) for x in stops_df['TRACT_ID']]
    for tract in tracts:
        values.append(dictionary[tract])
    return values
```

Assigning each stop the population of its tract.

```python
tract_impacted_pop_dict = get_pop_df_dict('impacted_pop')
impacted_pops = get_value_for_tract(tract_impacted_pop_dict)
```

```python
stops_df['impacted_pop'] = impacted_pops
stops_df
```

Assigning each tract the total employed population of its tract.

```python
tract_total_emp_dict = get_pop_df_dict('total_employed')
total_emp = get_value_for_tract(tract_total_emp_dict)
```

```python
stops_df['total_employed'] = total_emp
stops_df
```
