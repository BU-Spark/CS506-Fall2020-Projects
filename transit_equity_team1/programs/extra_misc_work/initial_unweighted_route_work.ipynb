```python
# imports
import pandas as pd
import numpy as np
from collections import Counter
```

```python
map_df = pd.read_csv('./output/baseline_map_info.csv')
map_df = map_df.drop(columns=['Unnamed: 0']) # redundant column with row number
map_df
```

```python
map_df['route_ids'] = map_df['route_ids'].astype("string")
print(map_df.dtypes)
```

```python
level_0 = map_df[map_df['income_level'] == 0] 
level_0
```

```python
print(level_0.dtypes)
```

```python
route_ids = [level_0.route_ids.apply(lambda x: x.split(',') if pd.notna(x) else None)]
```

```python
route_ids
```

```python
def get_route_ids(dataframe):
    all_route_ids = []
    for row in dataframe.itertuples():
        row_route_ids = row.route_ids
        if pd.notna(row_route_ids):
            route_id_list = row_route_ids.split(',')
            for route_id in route_id_list:
                all_route_ids.append(route_id)
    return all_route_ids

def get_counts(full_list):
    occurences = Counter(full_list)
    occurences_df = pd.DataFrame.from_dict(occurences, orient='index').reset_index()
    return occurences_df
```

```python
level_0_ids = get_route_ids(level_0)
level_0_ids
```

```python
level_0_df = get_counts(level_0_ids)
level_0_df
```

```python
level_1 = map_df[map_df['income_level'] == 1] 
level_1
```

```python
level_1_ids = get_route_ids(level_1)
level_1_df = get_counts(level_1_ids)
level_1_df
```

```python
level_1_df = level_1_df.rename(columns={'index':'route_id', 0:'level_1'})
level_1_df
```

```python
both_levels_df = level_0_df.merge(level_1_df, how='outer', on='route_id')
both_levels_df
```

```python
both_levels_df = both_levels_df.fillna(0)
both_levels_df
```

```python
both_levels_df['total'] = both_levels_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
both_levels_df
```

```python
cols = ['level_0','level_1','total']
both_levels_df[cols] = both_levels_df[cols].astype(int)
```

```python
both_levels_df['impacted_pop'] = 0
both_levels_df
```

```python
def get_pop():
    for row in map_df.itertuples():
        row_route_ids = row.route_ids
        if pd.notna(row_route_ids):
            route_id_list = row_route_ids.split(',')
            for route in route_id_list:
                both_levels_df.loc[both_levels_df.route_id == route, "impacted_pop"] += row.impacted_pop
    return
```
