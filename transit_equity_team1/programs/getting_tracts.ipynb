```python
import pandas as pd
from collections import Counter
```

```python
stops_complete_df = pd.read_csv('./output/stops_complete.csv')
stops_complete_df
```

```python
stops_complete_df.rename(columns = {'Income Lev':'income_level', 'Tract Id':'tract_id', 'Stop Id':'stop_id'}, inplace = True) 
stops_complete_df
```

```python
level_0 = stops_complete_df[stops_complete_df['income_level'] == 0] 
level_0
```

```python
# returns a list of tract_ids, including repeats, for the given dataframe
def get_tract_ids(dataframe): 
    all_tract_ids = []
    for row in dataframe.itertuples():
        all_tract_ids.append(row.tract_id)
    return all_tract_ids

# creates a dataframe containing each tract_id in the dataframe and the number of times it appears
def get_counts(full_list):
    occurences = Counter(full_list)
    occurences_df = pd.DataFrame.from_dict(occurences, orient='index').reset_index()
    return occurences_df
```

```python
level_0_ids = get_tract_ids(level_0)
level_0_ids
```

```python
level_0_df = get_counts(level_0_ids)
level_0_df
```

```python
level_0_df = level_0_df.rename(columns={'index':'tract_id', 0:'level_0'})
level_0_df
```

```python
level_1 = stops_complete_df[stops_complete_df['income_level'] == 1] 
level_1
```

```python
level_1_ids = get_tract_ids(level_1)
level_1_df = get_counts(level_1_ids)
level_1_df = level_1_df.rename(columns={'index':'tract_id', 0:'level_1'})
level_1_df
```

```python
both_levels_df = level_0_df.merge(level_1_df, how='outer', on='tract_id')
both_levels_df
```

```python
both_levels_df = both_levels_df.fillna(0)
both_levels_df
```

```python
both_levels_df['total'] = both_levels_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
both_levels_df
```

```python
cols = ['level_0','level_1','total']
both_levels_df[cols] = both_levels_df[cols].astype(int)
both_levels_df
```

```python
both_levels_df.sort_values('total', ascending=False)
```

```python
# returns a list of towns, including repeats, for the given dataframe
def get_town(dataframe): 
    all_towns = []
    for row in dataframe.itertuples():
        all_towns.append(row.Town)
    return all_towns
```

```python
level_0_towns = get_town(level_0)
level_0_town_df = get_counts(level_0_towns)
level_0_town_df = level_0_town_df.rename(columns={'index':'town', 0:'level_0'})
level_0_town_df
```

```python
level_1_towns = get_town(level_1)
level_1_town_df = get_counts(level_1_towns)
level_1_town_df = level_1_town_df.rename(columns={'index':'town', 0:'level_1'})
level_1_town_df
```

```python
all_towns_df = level_0_town_df.merge(level_1_town_df, how='outer', on='town')
all_towns_df
```

```python
all_towns_df = all_towns_df.fillna(0)
all_towns_df
```

```python
all_towns_df['total'] = all_towns_df.apply(lambda row: row.level_0 + row.level_1, axis=1) 
cols = ['level_0','level_1','total']
all_towns_df[cols] = all_towns_df[cols].astype(int)
all_towns_df
```

```python
def town_percentage(town_totals, town_low_income):
    percent_dict = {}
    for town in town_low_income.town:
        low_income_stops = town_low_income.loc[town_low_income['town'] == town, 'total'].values[0] # number of low-income stops
        all_stops = town_totals.loc[town_totals['town'] == town, 'counts'].values[0] # number of stops in total
        percent_low_income = low_income_stops/all_stops*100 # percentage of stops that are low-income for that town
        percent_dict[town] = percent_low_income
    return percent_dict
```

```python
town_percents = town_percentage(town_count_df, all_towns_df)
town_percents
```
