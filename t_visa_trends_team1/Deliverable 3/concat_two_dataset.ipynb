## Convert *date* column

```
months = [
          'January',
          'February',
          'March',
          'April',
          'May',
          'June',
          'July',
          'August',
          'September',
          'October',
          'November',
          'December',
          'Feb',
          'SEPT'
]

months_replacement = [
                      'JAN',
                      'FEB',
                      'MAR',
                      'APR',
                      'MAY',
                      'JUN',
                      'JUL',
                      'AUG',
                      'SEP',
                      'OCT',
                      'NOV',
                      'DEC',
                      'FEB',
                      'SEP'
]
```

```
for i in range(len(months)):
  mask_month=df2.date.map(lambda x: months[i] in x)
  for index, rows in df2[mask_month].iterrows():
    string = rows.date
    df2.date[index] = string.replace(months[i], months_replacement[i])
```

```
for index, rows in df2.iterrows():
  string = rows.date
  df2.date[index] = ((string.replace(" ", "")).replace(",","")).replace(".","")
```

```
df3 = pandas.concat([df, df2], ignore_index=True, sort=False)
```

```
for index, rows in df3.iterrows():
  string = rows.date
  string_length = len(df3.date[index])
  if string_length == 8:
    df3.date[index] = string[0:3] + '0' + string[3:]
```
