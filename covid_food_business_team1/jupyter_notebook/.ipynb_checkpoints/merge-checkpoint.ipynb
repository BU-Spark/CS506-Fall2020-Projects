```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Analyze Government Polices Impact

def filter(keyword, value, filepath, output_name):
    df = pd.read_csv(filepath)

    filtered = pd.DataFrame()

    df = df.dropna()
    for i in range(df.shape[0]):
        if value in df.loc[i][keyword]:
            filtered = filtered.append(df.loc[i])

    filtered.to_csv(output_name)


def remove_duplicate(filepath, keyword):
    """
    remove csv duplicate rows according to the keyword
    :param filepath:
    :param keyword:
    :return:
    """
    df = pd.read_csv(filepath)

    unique_keyword_list = {}
    res = pd.DataFrame()

    for i in range(df.shape[0]):
        if df.loc[i][keyword] in unique_keyword_list: continue
        else:
            res = res.append(df.loc[i])

    return res


if __name__ == '__main__':
    # filter boston area
    #filter('Neighborhood', 'Boston', 'small-business-relief-fund.csv', 'filtered_relief_fund.csv')

    # to lower case
    df = pd.read_csv('filtered_Small.csv')
    df['name_lower'] = df['Business Name'].str.lower()

    # merge with given file
    given_file = 'sg_09.csv'

    df_given = pd.read_csv(given_file)
    df_given['name_lower'] = df_given['location_name'].str.lower()
    pd.merge(df, df_given, on='name_lower').to_csv('merged_small.csv')

    # remove duplicate
    no_dup = remove_duplicate('merged_small.csv', 'placekey')
    no_dup.to_csv('merged_small_no_dup.csv')
    
    
```
